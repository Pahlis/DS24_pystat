### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 100)
y = x ** 2
```

```python
#object-oriented approach
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("object-oriented (y = x^2)")
ax.set_xlabel("label x")
ax.set_ylabel("label y")
```

```python
#pyplot aproach
plt.plot(x, y)

plt.title("Pyplot (y = x^2)")
plt.xlabel("label x")
plt.ylabel("label y")
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, ax= plt.subplots()
ax.scatter(random_data_x, random_data_y)
#ax.plot(x, y) I was unsure how to handle this input, so I did both. See below.

ax.set_title("Random data")
ax.set_xlabel("x")
ax.set_ylabel("y")
```

```python
fig, ax= plt.subplots()
ax.scatter(random_data_x, random_data_y)
ax.plot(x, y)

ax.set_title("Random data")
ax.set_xlabel("x")
ax.set_ylabel("y")
```

```python
fig, ax = plt.subplots()
ax.bar(names, values)

ax.set_title("Fruit data")
ax.set_xlabel("fruit")
ax.set_ylabel("value")
```
